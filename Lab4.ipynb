```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mushroom_data = pd.read_csv('https://raw.githubusercontent.com/kaopanboonyuen/SC310005_ArtificialIntelligence_2023s1/main/dataset/mushrooms_dataset.csv')
```

```python
mushroom_data.head()
```

Q1.Visualize the **distribution of classes** using a count plot in Seaborn.

```python
 #1
sns.countplot(x='class', data = mushroom_data)
```

Q2.Create a histogram of cap sizes **(`cap-surface`)**.

```python
#2
sns.histplot(mushroom_data['cap-surface'].dropna(),kde = True)
```

```python
sns.countplot(x='cap-surface', data=mushroom_data)
```

Q3.Generate a count plot to display the count of different **stalk shapes**.

```python
#3
sns.countplot(x='stalk-shape', data = mushroom_data)
```

Q4.Create a count plot to visualize the distribution of **gill sizes**.

```python
#4
sns.countplot(x='gill-size', data = mushroom_data)
```

Q5.Visualize the count of different **cap shapes** specifically for **edible**
mushrooms

```python
#5
edible_mushrooms = mushroom_data[mushroom_data['class']== 'e']
sns.countplot(x='class', data = edible_mushrooms)
```

```python
sns.countplot(x='cap-shape', data=edible_mushrooms)
```

Q6.Visualize the count of different **cap colors** specifically for **poisonous**
mushrooms

```python
#6
poisonous_cap_mushrooms = mushroom_data[mushroom_data['cap-color']== 'p']
sns.countplot(x='cap-color', data = poisonous_cap_mushrooms)
```

Q7.Visualize a Seaborn countplot with a new variable `'cap-surface-color'` by
combining `'cap-surface'` and `'cap-color'`.

```python
#7
mushroom_data['cap_surface_color'] = mushroom_data['cap-surface'] + mushroom_data['cap-color']
sns.countplot(x='cap_surface_color', data = mushroom_data)
```

Q8.Generate a **histogram** in Seaborn to visualize the distribution of spore **print

Q9.Visualize a count plot with a new variable, `'habitat-population'`, by
combining `'habitat'` and `'population'`.
